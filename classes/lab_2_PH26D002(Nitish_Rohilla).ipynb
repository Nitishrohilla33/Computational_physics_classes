## Problem 1
## Sum of elements in a list 

```python
def sum_of_elements(l):
    s=0
    n=len(l)
    for i in range(0,n):
        s=l[i]+s
    return s
```

```python

A = np.array([5,5,6])
print (
    f'sum of elements in list A = {sum_of_elements(A)}\n'
    f'Using numpy = {np.sum(A)}'
)
```

## Problem 2
## Minimum element and its index

```python
def minimum_element(l):
    n=len(l)
    minimum=l[0]
    minimum_index=0
    for i in range (n):
        if l[i]<minimum:
            minimum=l[i]
            minimum_index=i
    return minimum, minimum_index
```

```python
B=np.array([6,5,1,10],dtype = float)
print(minimum_element(B))
```

## Problem 3
## counting positive, negative and zero elements

```python
def counting(l):
    n=len(l)
    positive=0
    negative=0
    zero=0
    for i in range(n-1):
        if l[i]<0:
            negative=negative+1
        elif l[i]==0:
            zero=zero+1
        elif l[i]>0:
                positive=positive+1
    
    return negative, zero, positive
```

```python
C=[0,0,1,-1,5,5]
print(counting(C))
```

## Problem 4
## Ascending and descending order of an arrey

```python
def minimum_element(l):
    n=len(l)
    minimum=l[0]
    minimum_index=0
    for i in range (n):
        if l[i]<minimum:
            minimum=l[i]
            minimum_index=i
    return minimum, minimum_index
def ascending(l):
    l = list(l)
    n=len(l)
    l_new = []
    for i in range(n):
        mini, index = minimum_element(l)
        #print(mini,index)
        del l[index]
        l_new.append(mini)
    return l_new         
```

```python
D=[1,845,5,687,354,5465]
print(ascending(D))
```

## Problem 5
## Motion of mass on a spring

```python
#function to compute x(t) for a given list of time values t
def displacement(a,w,t):
    cos_t=np.cos(w*t)
    x=a*cos_t
    return x

t=np.linspace(0,20,1000)
a=1
w=.25
y=displacement(a,w,t)
plt.plot(t,y)
plt.grid()
plt.xlabel("Time (radians)")
plt.ylabel("Displacement")
plt.title("Simple Harmonic Motion")
plt.show()
```

### Problem 6
### Bouncing projectile motion

```python
# constants
g = -9.81 # m/s^2

# initial condition
x,y = 0,0 # m

# velocity and angle
v = 10 # m/s
theta = 45 # degree
theta = theta*np.pi/180 # radian
vx,vy = v*np.cos(theta),v*np.sin(theta) 

# steps and time 
steps = 1000000
dt = 0.000007
t = 0 # seconds

# trajactories 
x_traj,y_traj,t_traj = [x],[y],[t]

for i in range(steps):
    x += vx*dt
    y += vy*dt
    vy += g*dt
    t += dt
    
    if y <= 0:
        y = 0
        
        # current speed before bounce
        speed = np.sqrt(vx**2 + vy**2)
        
        # current angle
        theta = np.arctan2(vy, vx)
        
        # small random perturbation
        theta += 0.05 * (2*np.random.rand() - 1)   # adjust 0.05 as needed
        
        # coefficient of restitution
        e = 0.8
        
        # new velocity after bounce
        vx = speed * np.cos(theta)
        vy = abs(e * speed * np.sin(theta))  # upward after bounce

    # append trajactories
    x_traj.append(x)
    y_traj.append(y)
    t_traj.append(t)

plt.plot(x_traj,y_traj)
plt.grid()
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Height (m)")
plt.title("Projectile Motion with Bouncing")
plt.show()
```
