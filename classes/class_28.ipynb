```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
A = np.array([[2,1],
              [1,2]])
max_iter = 15

u = np.array([[4],[16]])

for i in range(max_iter):
    v = A@u
    norm = np.linalg.norm(v)

    u = v/norm

    lam = u.T@A@u/(u.T@u)
    #np.set_printoptions(precision=4, suppress=True)    
    print(
        f'Eigenvalue = {lam[0]}  '
        f'Eigenvector = {u[0,0], u[1,0]}'
        ) 
```
