### to calculate the value of pi 
- imagine a circle imbeded in a square. Use Monte Carlo method simulate throughing dart at random points on that figure and calculate the value of pi.

```python
# define square 
N = 1000
x = 2*np.random.random(N) - 1
y = 2*np.random.random(N) - 1
plt.plot(x,y, color = 'red', marker = '.', linestyle = '')
plt.xlabel('x axis', font = 'times new roman')
plt.ylabel('y axis', font = 'times new roman')
plt.title('Throughing dart at random points', font = 'times new roman')
#plt.grid()
```

```python
N = 90000
x = 2*np.random.random(N) - 1
y = 2*np.random.random(N) - 1
count = 0

for i in range(N):
    if x[i]**2+y[i]**2 <= 1:
        count = count+1
print(4*count/N)        
```

### Difusion of ink in water
- Brownian motion

```python
def random_walk(number_of_steps):
    x = 0
    for step in range(number_of_steps):
        if np.random.random()<0.5:
            x = x+1
        else:
            x = x-1

    return x
```

```python
trial = 100000
Nsteps = 10
p = []
for t in range(trial):
    xfinal = random_walk(Nsteps)
    p.append(xfinal)
```

```python
p = np.array(p)
print(np.mean(p))
print(np.mean(p**2))
```

```python
trial = 10
for i in range(trial):
    number_of_steps = 100
    x = 0
    pos = []
    for step in range(number_of_steps):
        if np.random.random()<0.5:
            x = x+1
        else:
            x = x-1
        pos.append(x)
    
    step = np.arange(number_of_steps)
    plt.plot(step, pos)
    plt.xlabel('Number of collisions', font = 'times new roman', size = 20)
    plt.ylabel('Position of the particles', font = 'times new roman', size = 20)
    plt.title('Dispersion of ink in water', font = 'times new roman', size = 24)
    #plt.grid()
```
