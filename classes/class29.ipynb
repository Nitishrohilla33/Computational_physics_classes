```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Differential equation

```python
def f(x,y):
    return y
```

```python
def euler(xf, h, x0 = 0, y0 = 1):
    x = np.arange(x0,xf,h)
    y = np.zeros(len(x))
    x[0] = 0
    y[0] = y0 

    for i in range(len(x)-1):
        y[i+1] = y[i] + h*f(x[i],y[i])

    return x,y    
```

```python
def rk4(xf, h, x0 = 0, y0 = 1):
    x = np.arange(x0,xf,h)
    y = np.zeros(len(x))
    x[0] = 0
    y[0] = y0 

    for i in range(len(x)-1):
        k1 = f(x[i],y[i])
        k2 = f(x[i]+h/2,y[i]+k1*h/2)
        k3 = f(x[i]+h/2,y[i]+k2*h/2)
        k4 = f(x[i]+h,y[i]+k3*h)
        y[i+1] = y[i] + h/6*(k1+2*(k2+k3)+k4)

    return x,y
```

```python
x,y = euler(xf = 3, h = 0.2)
x1,y1 = rk4(xf = 3, h = 0.2)
plt.plot(x,y, marker = '.', color = 'b', linestyle = '--', label = 'Eular')
plt.plot(x1,y1, marker = '.', color = 'g', linestyle = '--', label = 'RK4')
plt.plot(x, np.exp(x), color = 'black',label = 'Exact')
plt.legend()
plt.show()
```
