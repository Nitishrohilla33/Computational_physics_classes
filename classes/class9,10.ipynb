```python
birthday = np.random.randint(1, 366, size = 23)
birthday = list(birthday)
n = len(birthday)
y = 0
same = 0
for m in range (1000):    
    for i in range(n):
        for j in range(i+1, n-1):
            if birthday[j] == birthday[i]:
                same = same + 1
    if same >= 1:
        y = y+1
print(y/1000)
```
