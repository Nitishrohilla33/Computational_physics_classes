```python
# plot for rotations
theta = np.pi*90/180 # angles
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.cos(theta), np.sin(theta)]])
v = np.array([[2],
              [2]])
```

```python
plt.plot([0, v[0, 0]], [0, v[1, 0]], marker = '.', color = 'red')
plt.plot([0, (A@v)[0, 0]], [0, (A@v)[1, 0]], marker = '.', color = 'b', linestyle = '--')
plt.axhline(0, linestyle = ':')
plt.axvline(0, linestyle = ':')
plt.grid()
```

```python
theta = np.pi*60/180 # angles
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.cos(theta), np.sin(theta)]])

x = np.arange(-5, 5.1, 0.5)
y = np.arange(-5, 5.1, 0.5)

for i in range(len(x)):
    for j in range(len(y)):
        v = np.array([[x[i]],
                      [y[j]]])
        plt.plot([0, v[0, 0]], [0, v[1, 0]], marker = '.', color = 'red', linestyle = ' ')
        plt.plot([0, (A@v)[0, 0]], [0, (A@v)[1, 0]], marker = '.', color = 'b', linestyle = ' ', zorder = 300)

plt.axhline(0, linestyle = ':')
plt.axvline(0, linestyle = ':')
plt.axis('equal')
#plt.grid()
```
