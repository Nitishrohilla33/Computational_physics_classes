### Let's solve Monty Hall problem

```python
trials = 100000
doors = [0, 1, 2]
prizes = ['goat', 'goat', 'car']
switch = True
wins = 0
for n in range(trials):    
    choice = np.random.choice(doors)
    
    
    # find the door that Monty will open
    for i in range (3):
        if i != choice and prizes[i] != 'car':
            open_door = i
    
    if switch == True:
        for j in range(3):
            if j != choice and j != open_door:
                choice_new = j
        if prizes[choice_new] == 'car':
            wins += 1        
    else:
        if i == choice and prizes[i] == 'car':
            wins += 1
            
                   
    #print(wins)        
    #print('dooe opened:', open_door, prizes[open_door], 'choice was:', choice, prizes[choice], 'switch:', choice_new, prizes[choice_new])      
print('probability of winning = ', wins/trials)    
```
