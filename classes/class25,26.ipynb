```python
import numpy as np
import matplotlib.pyplot as plt
import math
import emcee
```

```python
def gauss_elimination(A):
    n = A.shape[0]
    for i in range (n):
        for k in range (i+1,n):
            m = A[k,i]/A[i,i]
            A[k,: ] = A[k,: ] - m*A[i,: ]
    return A
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
print(A)
print(gauss_elimination(A))
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
b = np.array([[9],[-10],[-2],[5]], dtype = float)
B = np.hstack((A,b))
print(B)
print(gauss_elimination(B))
```

```python
def sol(A,b):
    n = A.shape[0]
    s = gauss_elimination(np.hstack((A,b)))
    solution = np.zeros((n,1)) 
    for i in range(n-1,-1,-1):
        solution[i] = s[i,n]
        for j in range(i+1,n):
            solution[i] = solution[i]-solution[j]*s[i,j]
        solution[i] = solution[i]/s[i,i]
    
    return solution    
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
b = np.array([[9],[-10],[-2],[5]], dtype = float)
print(sol(A,b))
```

# Gauss-Jordan elimination

```python
def gauss_Jordan_elimination(A):
    n = A.shape[0]
    for i in range (n):
        for k in range (n):
            if i != k:
                A[k,: ] = A[k,: ] - A[i,: ]*A[k,i]/A[i,i]
    return A
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
print(gauss_Jordan_elimination(A))
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
b = np.array([[9],[-10],[-2],[5]])
B = np.hstack((A,b))
Z = gauss_Jordan_elimination(B)
print(Z)
Z = np.array([Z[i,:]/Z[i,i] for i in range(Z.shape[0])])
print (Z[:,4:])
```

```python
def gauss_Jordan_solution(A):
    n = A.shape[0]
    for i in range (n):
        A[i] = A[i]/A[i,i]
        for k in range (n):
            if i != k:
                A[k,: ] = A[k,: ] - A[i,: ]*A[k,i]
    return A
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
b = np.array([[9],[-10],[-2],[5]])
B = np.hstack((A,b))
print(gauss_Jordan_solution(B))
print(gauss_Jordan_solution(B)[:,4:])
```

```python
def inverse(A):
    n = A.shape[0]
    I = np.eye(A.shape[0])
    A = np.hstack((A,I))
    for i in range (n):
        A[i] = A[i]/A[i,i]
        for k in range (n):
            if i != k:
                A[k,: ] = A[k,: ] - A[i,: ]*A[k,i]
    return A[:,A.shape[0]:]
```

```python
A = np.array([[2, 1, -1, 1], [-3, -1, 2, 1], [-2, 1, 2, 1], [5, 2, 6, -5]],dtype=float)
inv = inverse(A)
np.set_printoptions(precision=2)
print(inv)
```

```python
A = np.array([[2,1],
              [1,2]])
print(np.linalg.inv(A))
inv = inverse(A)
print(inv)
```
