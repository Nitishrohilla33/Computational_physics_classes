### Question 2
- (a)

```python
# value of slope.
def least_fit(x, y):
    n = len(x)
    XY,X,Y,X2 = 0,0,0,0
    for i in range(n):
        XY += x[i]*y[i]
        X += x[i]
        Y += y[i]
        X2 += x[i]**2
    return (n*XY - X*Y)/(n*X2-(X)**2), (X2*Y-XY*X)/(n*X2-X**2)
```

```python
m,c = least_fit(x,y)
```

```python
print('m = ', m)
print('c = ', c)
```

```python
# residual
x = np.array([-5.0 , -4.5, -4.0 , -3.5, -3.0 , -2.5, -2.0, -1.5, -1.0, -0.5, 0.0 , 0.5,
1.0 , 1.5, 2.0 , 2.5, 3.0 , 3.5, 4.0, 4.5, 5.0 ])
y = np.array([-19.0, -14.7, -13.9, -8.9, -12.4, -15.6, -10.7, -7.1, -9.7, -6.0, 2.2,
-2.5, -0.8, 3.4, 1.2, 7.4, 3.8, 12.5, 11.8, 10.4, 16.9])
```

```python
residual = []
for i in range(len(x)):
    m,c = least_fit(x, y)
    r = y[i]-m*x[i]-c
    residual.append(r)

plt.scatter(x, residual, color = 'g', edgecolor = 'k')
plt.xlabel('x')
plt.ylabel('residual')
```

```python
print('The sum of residuals = ', np.sum(residual))
```

```python
x = np.array([-5.0 , -4.5, -4.0 , -3.5, -3.0 , -2.5, -2.0, -1.5, -1.0, -0.5, 0.0 , 0.5,
1.0 , 1.5, 2.0 , 2.5, 3.0 , 3.5, 4.0, 4.5, 5.0 ])
y = np.array([-19.0, -14.7, -13.9, -8.9, -12.4, -15.6, -10.7, -7.1, -9.7, -6.0, 2.2,
-2.5, -0.8, 3.4, 1.2, 7.4, 3.8, 12.5, 11.8, 10.4, 16.9])
```

```python
m_vals = np.linspace(3, 3.5, 200)
c_vals = np.linspace(-3, -2, 200)
S = np.zeros((len(c_vals), len(m_vals)))
for i in range(len(m_vals)):
    for j in range(len(c_vals)):
        y_model = m_vals[i] * x + c_vals[j]
        S[j, i] = np.sum((y - y_model)**2)
plt.contour(m_vals, c_vals, S, 50)
plt.show()
```

```python
idx = np.unravel_index(np.argmin(S),S.shape)
intercept = c_vals[idx[0]]
slope = m_vals[idx[1]]
print(
    f'Best slope = {slope}\n'
    f'Best intercept = {intercept}'
)
xfit = np.linspace(np.min(x),np.max(x)+0.1, 1000)
yfit = slope*xfit+intercept
plt.plot(xfit, yfit, color = 'g')
plt.scatter(x,y, marker = 'o',color = 'b')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(m_vals, c_vals, S)
plt.show()
```

### question 3

```python
X = [1, 2, 3, 4, 5, 6, 8, 10, 12, 15, 17]
Y = [2.1, 6.5, 12.8, 21.6, 39.4, 46.9, 77.8, 137.2, 167.5, 291.3, 319.5]
plt.scatter(X, Y, color = 'c', edgecolor = 'k')
```

```python
n = 200
m_vals = np.linspace(1.5,2,n)
c_vals = np.linspace(1,2,n)
s = np.zeros((len(m_vals),len(c_vals)))
for i in range(len(m_vals)):
    for j in range(len(c_vals)):
        y_model = c_vals[j]*(X**m_vals[i])
        s[i,j] = np.sum((Y-y_model)**2)
plt.contour(m_vals,c_vals,s,50)
plt.show()
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(m_vals, c_vals, s)
plt.show()
```

```python
idx = np.unravel_index(np.argmin(s),s.shape)
intercept = c_vals[idx[1]]
slope = m_vals[idx[0]]
xfit = np.linspace(min(X),max(X),n)
yfit = intercept*(xfit**slope)
plt.plot(xfit,yfit,label = 'best fit')
plt.scatter(X, Y, color = 'c', edgecolor = 'k')
plt.legend()
plt.show()
```
