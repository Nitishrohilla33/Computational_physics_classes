```python
import numpy as np
import matplotlib.pyplot as plt
```

### Linear fitting

```python
x = np.arange(-5,5.1,0.5)

y = 3*x

plt.scatter(x, y, facecolor = 'cyan', edgecolor = 'k', label = 'data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend
```

```python
y = 3*x
y = [y[i]-3+6*np.random.random() for i in range(len(y))]

plt.scatter(x, y, facecolor = 'cyan', edgecolor = 'k', label = 'data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend
```

```python
y = 3*x
np.random.seed(42)
y = y + np.random.normal(0,2,len(y))
plt.scatter(x, y, facecolor = 'cyan', edgecolor = 'k', label = 'data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend
```

```python
slopes = np.arange(0, 6, 0.1)

lsq = []
for m in slopes:
    f = np.sum((m*x-y)**2)
    lsq.append(f)
    

plt.plot(slopes, lsq)
plt.xlabel('slope = m')
plt.ylabel('least square')
plt.legend
```

```python
idx = np.argmin(lsq)
result = slopes[idx]
print(result)
```
