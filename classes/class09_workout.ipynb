```python
def is_birthday_same(number_of_students):
    n = number_of_students # No. of birthdays
    birthdays = np.random.randint(1, 366, size = n)
    shared = False
    for i in range(n):
        for j in range(i+1, n):
            if birthdays[i] == birthdays[j]:
                shared = True
                #print('shared', i,j)
        if shared == True:
            #print('shared after break', i,j)
            break
    return shared            
```

```python
class_size = 23
trials = 100000
count = 0
for c in range(trials+1):
    shared_bday = is_birthday_same(number_of_students=class_size)
    if shared_bday == True:
        count = count + 1       
print('probebility = ', count/trials)
```

```python
class_size = np.arange(2, 100)
p = []
for c in class_size:
    count = 0
    trial = 10000
    for t in range(trial):
        shared_bday = is_birthday_same(number_of_students=c)
        if shared_bday == True:
            count = count + 1 
    p.append(count/trial)
```
